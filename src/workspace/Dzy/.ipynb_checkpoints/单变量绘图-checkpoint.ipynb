```python
import sys
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
from pylab import mpl
```

```python
d1 = pd.read_csv("./data/process/Temp_predict_06.csv",index_col=0)
d2 = pd.read_csv("./data/process/Temp_predict_07.csv",index_col=0)
df = d1.append(d2)
df.index = pd.to_datetime(df.index)
df = df.sort_index()
# df['date'] = df.index.date
# df['month'] = df.index.strftime("%Y-%m")
df['label'] = df['label'].map({0:u'正常状态',1:u'异常状态'})
```

```python
plt.figure(figsize=[20,5])
sns.set_style('whitegrid')
plt.rcParams['font.family'] = 'SimHei' 
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
col = 'slope'

plt.subplot(1,2,1)
ax1 = sns.boxplot(data = df,x= 'label',y = col,sym='')
ax1.set_title('特征%s'%col)
ax1.set_xlabel('')
ax1.set_ylabel(u'温度(℃)')

plt.subplot(1,2,2)
ax2 = sns.distplot(df[df.label=='正常'][col],label = '正常状态')
ax2 = sns.distplot(df[df.label=='异常'][col], label = '异常状态')
ax2.set_xlabel('')
ax2.set_xlim([-0.05,0.05])
# ax2.set_ylabel(u'温度(℃)')
ax2.set_ylabel('')
plt.legend(loc = 'best')
plt.show()
```

```python
plt.figure(figsize=[30,5])
sns.set_style('whitegrid')
plt.rcParams['font.family'] = 'SimHei' 
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
sns.boxplot(data = df,x= 'date',y = 'TE_1003',hue='label',sym='')
plt.title('平均温度日均差异性对比')
plt.xlabel('')
plt.ylabel(u'温度(℃)')
plt.legend()
plt.xticks(rotation=20)
```

```python
def plot_temp_overview(df, alpha = 0.8,col = 'temp',title=u'特征1: temp'):
    '''
    df: DataFrame数据
    alpha: 设定线条的透明度，默认为0.8
    
    '''
    sns.set_style('darkgrid')
    plt.rcParams['font.family'] = 'SimHei' 
    plt.rcParams['font.size'] = 20
    plt.rcParams['axes.unicode_minus'] = False
    plt.figure(figsize=[20,5])
#     ax1.xaxis.set_major_locator(ticker.MultipleLocator(280))
#     ax.xaxis.set_major_locator(matplotlib.dates.DayLocator(interval=2))
    plt.plot(df['label'], label = u'工况状态', alpha = alpha)
    plt.plot(df['Logistic'], label = u'异常风险值', alpha = alpha)
#     plt.plot(df['TE_1003c'], label = u'浓缩段温度C', alpha = alpha)
    
#     plt.plot(df['TE_1003'], label = '浓缩段平均温度', alpha = alpha, color = '#8600FF')
    
#     df['index'] =df.index
#     temp = df.pivot_table(index= "index",columns = 'label',values = col)
#     plt.plot(temp['正常'], label = '正常状态', alpha = alpha, color = 'green')
#     plt.plot(temp['异常'], label = '异常状态', alpha = alpha, color = 'red')
    t = pd.Series([0.60,0.55],index = [df.index.min(),df.index.max()])
    plt.plot(t,":",color = 'blue',label = '预警阈值')
#     plt.xlabel('Datetime')
#     for i in ['2019-06-30 21:45:00', '2019-06-30 22:03:00',
#                '2019-06-30 22:18:00', '2019-06-30 22:33:00','2019-06-30 22:49:00',
#                '2019-06-30 23:03:00', '2019-06-30 23:19:00',
#                '2019-06-30 23:32:00', '2019-06-30 23:45:00']:
#         t = pd.Series([57,63],index = pd.to_datetime([i,i]))
#         plt.plot(t,":",color = '#8600FF')
#     plt.title(title)
    plt.xticks(rotation=20)
    plt.ylabel('')
    plt.legend(loc = 'upper left')
    return plt.show()
```
